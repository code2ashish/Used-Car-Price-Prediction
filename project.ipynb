```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import statsmodels.api as sm #stastics calculation
```

```python
df=pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
```

```python
df.head()
```

```python
df["Age"]=2022-df.year
df.drop(["name","year"],axis=1,inplace=True)
y=df.selling_price
df.drop(["selling_price"],axis=1,inplace=True)
df
```

```python
y
```

```python
f=pd.get_dummies(df.fuel,prefix="Fuel_Type")
s=pd.get_dummies(df.seller_type,prefix="Seller_Type")
t=pd.get_dummies(df.transmission,prefix="Transmission")
o=pd.get_dummies(df.owner,prefix='owner')
df.drop([ 'fuel', 'seller_type','transmission','owner'],axis=1,inplace=True)
```

```python
X=pd.concat([df,f,s,t,o],axis=1)
X
```

```python
from xgboost import XGBRegressor
model=XGBRegressor()
```

```python
model.fit(X,y)
```

```python
y_pred=model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,
print('MSE:', mean_absolute_error(y,y_pred))
print('R Squre:',r2_score(y,y_pred))
```

```python

from sklearn.linear_model import LinearRegression,LogisticRegression,Ridge,Lasso,PoissonRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
model_list=[LinearRegression(),LogisticRegression(),Ridge(),Lasso(),PoissonRegressor(),SVR()]
for model in model_list:
    print(str(model))
    model.fit(X,y)
    y_pr=model.predict(X)
    print('MSE:', mean_absolute_error(y,y_pr))
    print('R Squre:',r2_score(y,y_pr))
    print('*'*50)

```
